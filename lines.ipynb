```python
from overturemaps import core
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# Manhattan
bbox = (-74.047285, 40.682944, -73.9067, 40.8776)

# olympia
bbox = (-123.0660, 47.0290, -122.8500, 47.1000)

```

```python
# download the data for segments (transportation)
# check out https://docs.overturemaps.org/guides/transportation/
gdf = core.geodataframe("segment", bbox=bbox)
gdf.head(5)
```

```python
gdf[gdf['class']=="pedestrian"].head()
```

```python
# Plot with reduced linewidth

fig, ax = plt.subplots(figsize=(12, 12))
gdf.clip(bbox).plot(ax=ax, color="black", linewidth=0.7, alpha=.5)  # Adjust linewidth to desired thickness
ax.set_title("Transportation, Olympia, WA, USA")
plt.show()
```
